```python
from matplotlib import pyplot as plt
import numpy as np
results=np.loadtxt("SBM_Lighton_Adjacency.csv")
mult_factor=np.append(np.arange(1,1.1,0.02),np.arange(1.1,1.8,0.2))
nodes_num=range (5,8)

results =results.reshape(3,10,2).mean(axis=-1)
print(results.shape)
for (ind,nodes) in enumerate(nodes_num):
    plt.plot(mult_factor,100*results[ind].flatten(),linewidth=3,label="size= "+str(nodes))

plt.xlabel("Inter-classes similarity parameter")
plt.ylabel("Test accuracy")
plt.title("Lighon_adjacency different graphlet sizes and classes similarity")
plt.grid()
plt.legend()
plt.savefig('Lighton random features with adjacency matrix.png')
```
