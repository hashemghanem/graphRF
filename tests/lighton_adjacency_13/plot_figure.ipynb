```python
from matplotlib import pyplot as plt
import numpy as np
results=np.loadtxt("Lighton_adjacency_SBM_uniform_samp_num.csv")
mult_factor=1.5
print(results.shape)
nodes_num=range (4,7)
num_expe=1
features_num=300
samples_num=range (50,1007,50)


results =results.reshape(len(nodes_num),len(samples_num),num_expe)
results=results.mean(axis=-1)
print(results.shape)
colors=['r','b','g', 'k','c','m']
for (ind,nodes) in enumerate(nodes_num):
    plt.plot(samples_num,100*results[ind].flatten(),linewidth=1.8,label="size= "+str(nodes), color=colors[ind])

plt.xlabel(" Number of graphlet samples")
plt.ylabel("Test accuracy")
plt.title("Lighton_adjacency different graphlet sizes and samples number")
plt.grid()
plt.legend()
plt.savefig('Lighton random features with adjacency matrix.png')
```
