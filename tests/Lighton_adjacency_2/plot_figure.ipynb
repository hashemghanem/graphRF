```python
from matplotlib import pyplot as plt
import numpy as np
results=np.loadtxt("SBM_Lighton_Adjacency.csv")
mult_factor=np.arange(1,1.06,0.005)
print(results.shape)
plt.plot(mult_factor,results,linewidth=3)

plt.xlabel("Inter-classes similarity parameter")
plt.ylabel("Test accuracy")
plt.title("Lighon_adjacency different levels of classes similarity")
plt.grid()
plt.legend(["Graphlet size:6\n random features: 500"])
plt.savefig('Lighton random features with adjacency matrix.png')
```
