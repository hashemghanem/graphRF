```python
from matplotlib import pyplot as plt
import numpy as np
results=np.loadtxt("DD_Lighton_Adjacency0.csv")
ss=np.loadtxt("DD_Lighton_Adjacency.csv")
feat_axis=[10, 25, 50, 100, 250, 500, 1000,5000] 
print(len(feat_axis))
print(ss.shape)
a=np.append(results.reshape(len(feat_axis),1),ss.reshape(len(feat_axis),2),axis=1)
print(a.shape)
nodes_num=7
samples_num=4000
plt.semilogx(feat_axis,100*a.mean(axis=1),linewidth=2,label="Avg accuracy of different experiments")
plt.scatter(np.sort(feat_axis*3),100*a,color='r',label="accurace of individual experiments")

plt.xlabel("Number of random features")
plt.ylabel("Test accuracy")
plt.title("Lighon_adjacency with different numbers of random features")
plt.grid()
plt.legend()
#plt.legend(["samples_num: {}\n graphlet_size:{}".format(samples_num,nodes_num)])
plt.savefig('DD-Lighton random features with adjacency matrix.png')
```
