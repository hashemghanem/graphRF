```python
from matplotlib import pyplot as plt
import numpy as np
results=np.loadtxt("SBM_Lighton_Adjacency.csv")
mult_factor=1+np.linspace(0.2,2,9)
nodes_num=range (3,6)
num_expe=1

results =results.reshape(len(nodes_num),mult_factor.shape[0],num_expe).mean(axis=-1)
for (ind,nodes) in enumerate(nodes_num):
    plt.plot(mult_factor,100*results[ind].flatten(),linewidth=1.8,label="size= "+str(nodes))

plt.xlabel("(Inter-classes similarity parameter )")
plt.ylabel("Test accuracy")
plt.title("Lighon_adjacency different graphlet sizes and classes similarity")
plt.grid()
plt.legend()
plt.savefig('Lighton RF Ind_RW with adjacency matrix.png')
```
