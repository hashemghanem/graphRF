```python
from matplotlib import pyplot as plt
import numpy as np
results=np.loadtxt("SBM_Lighton_Adjacency_uniformsampling.csv")
results_=np.loadtxt("Lighton_adjacency_Uniform_sampling_.csv")
results_=np.append(results_,results_[results_.shape[0]-1])
#results=np.concatenate((results,results_),axis=0)
mult_factor=1+np.linspace(0.2,2,8)
nodes_num=range (3,7)
num_expe=3

results =results.reshape(len(nodes_num),mult_factor.shape[0],num_expe)
results_=results_.reshape(len(nodes_num),mult_factor.shape[0],1)
results=np.concatenate((results,results_),axis=-1)
results=results.mean(axis=-1)
print(results.shape)
colors=['r','b','g', 'k','c','m']
for (ind,nodes) in enumerate(nodes_num):
    plt.plot(mult_factor,100*results[ind].flatten(),linewidth=1.8,label="size= "+str(nodes), color=colors[ind])

plt.xlabel("(Inter-classes similarity parameter )")
plt.ylabel("Test accuracy")
plt.title("Lighon_adjacency different graphlet sizes and classes similarity")
plt.grid()
plt.legend()
plt.savefig('Lighton random features with adjacency matrix.png')
```
