```python
from matplotlib import pyplot as plt
import numpy as np
OPU_time=np.loadtxt('OPU_time.csv')
GS_time=np.loadtxt('GS_time.csv')
EIG_time=np.loadtxt('EIG_time.csv')
Gk_time=np.loadtxt('Gk_time.csv')
nodes_num=range(3,36)
plt.plot(nodes_num,OPU_time,linewidth=2,marker='o', markersize=12,markerfacecolor="None",label='OPU')
plt.plot(nodes_num,GS_time,linewidth=2,marker='^', markersize=12,markerfacecolor="None",label='Gaussian')
plt.plot(nodes_num,Gk_time,linewidth=2,marker='v', markersize=12,markerfacecolor="None",label='Graphlet')
plt.plot(nodes_num,EIG_time,linewidth=2,marker='x', markersize=12,markerfacecolor="None",label='Gaussian+EIG')

plt.xlabel('Graphlet size')
plt.ylabel('Computation time (SEC)')
plt.grid()
plt.legend()
plt.yscale('log')
plt.savefig('Computation_time.pdf')
plt.show()    
```
